## Arrays in loops

```python
arr3d = np.array([
    [[8,7], [6,5]], 
    [[4,3], [2,1]],
    [[10,11], [12,13]]
])
```

```python
for i in arr3d:
    print(i)
```

```python
for blueberry in arr3d:
    for raspberry in blueberry:
        print(raspberry)
```

```python
for i in arr3d:
    for x in i:
        for g in x:
            print(g * .75)
```

```python
for i in arr3d:
    for x in i:
        print(sum(x))
```

## Concatenating and Stacking Arrays

```python
# array concatenation

arr_a = np.array([1, 2, 3])
arr_b = np.array([4, 5, 6])

arr_c = np.concatenate((arr_a, arr_b))  # concatenation results in the same number of dimensions as the starting arrays
print(arr_c)
```

```python
arr_d = np.stack((arr_a, arr_b))   # stack adds a new dimension to the resulting array
print(arr_d)
```

```python
print(arr_d.ndim)
print(arr_d.shape)
```

```python
combo_d = np.concatenate((arr_d, arr_d), axis = 0)
print(combo_d)
```

```python
print(combo_d.ndim)
print(combo_d.shape)
```

```python
combo_d2 = np.concatenate((arr_d, arr_d), axis = 1)
print(combo_d2)
```

```python
# using a practical example

# say I've got 2 sets of exam grades, from a original exam date and then a later makeup for some students who needed it
# exam_scores = [.78, .89, .92, .82, .99]
# exam_scores_makeup = [.90, .85]
# exam_all = np.concatenate((np.array(exam_scores), np.array(exam_scores_makeup)))

exam_scores = np.array([.78, .89, .92, .82, .99])
exam_scores_makeup = np.array([.90, .85])
exam_all = np.concatenate((exam_scores, exam_scores_makeup))

print(exam_all)
```

```python
project_scores = np.array([.88, .91, .97, .85, .96, .89, .87])

gradebook = np.stack((exam_all, project_scores))
print(gradebook)
```

### Working with hstack, vstack, dstack

```python
first_trio = np.array([1, 2, 3])
second_trio = np.array([4, 5, 6])

print(first_trio.ndim)
print(first_trio.shape)
```

```python
six_with_hstack = np.hstack((first_trio, second_trio))  # stacks each 1D array "horizontally" end-to-end
print(six_with_hstack)
```

```python
six_with_vstack = np.vstack((first_trio, second_trio))  # stacks each 1D array on top of each other "vertically"
print(six_with_vstack)
```

```python
six_with_dstack = np.dstack((first_trio, second_trio))  # stacks each 1D array in a new third dimension for depth
print(six_with_dstack)
```

```python
print(six_with_dstack.ndim)
```

```python
print(six_with_dstack.shape)
```

```python
first_quartet = np.array([1, 2, 3, 10])
second_quartet = np.array([4, 5, 6, 20])

print("using hstack:")
print(np.hstack((first_quartet, second_quartet)))
print(f'\nusing vstack:')
print(np.vstack((first_quartet, second_quartet)))
print(f'\nusing dstack:')
print(np.dstack((first_quartet, second_quartet)))
```

```python
a = six_with_vstack
b = six_with_vstack
c = six_with_vstack

print(np.hstack((a, b, c)))   # hstack will give you a resulting array with the same # of dimensions
```

```python
print(np.dstack((a, b)))     # results in 3D array
print(np.dstack((a, b)).ndim)
print(np.dstack((a, b)).shape)
```

```python
# what do we get if we use hstack with 3D arrays?

print(np.hstack((np.dstack((a, b)), np.dstack((a, b)))))
```

```python
# what do we get if we use vstack with 3D arrays?

print(np.vstack((np.dstack((a, b)), np.dstack((a, b)))))
```

```python
# what do we get if we use dstack with 3D arrays?

print(np.dstack((np.dstack((a, b)), np.dstack((a, b)))))
```

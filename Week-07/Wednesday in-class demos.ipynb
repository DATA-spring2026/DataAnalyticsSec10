```python
# .fillna() method can be used to provide a replacement value for NaN values
```

```python
# replacing values using a loop

for i in ice_cream.index:
    if ice_cream.loc[i, 'Flavor'] == 'Vanilla':
        ice_cream.loc[i, 'Price Per Scoop'] = 2.75
    elif ice_cream.loc[i, 'Price Per Scoop'] == "$2.50":
        ice_cream.loc[i, 'Price Per Scoop'] = 3.0
    elif ice_cream.loc[i, 'Price Per Scoop'] == "$2.75":
        ice_cream.loc[i, 'Price Per Scoop'] = 3.50
    elif ice_cream.loc[i, 'Price Per Scoop'] == "$3.00":
        ice_cream.loc[i, 'Price Per Scoop'] = 3.75

ice_cream
```

## Working with aggregation functions

```python
ice_cream
```

```python
ice_cream['Scoops Sold'].sum()
```

```python
ice_cream['Scoops Sold'].aggregate(["sum", "mean", "max", "min"])
```

```python
ice_cream[["Scoops Sold", "Price Per Scoop"]].aggregate(["sum", "mean"]).round(2)
```

```python
ice_cream.agg({
    "Scoops Sold": ["sum", "mean"],
    "Price Per Scoop": ["mean", lambda x:x.max() - x.min()]
})
```

```python
# using groupby

ice_cream.groupby(['Price Per Scoop']).sum(['Scoops Sold'])
```

```python
ice_cream.groupby(['Price Per Scoop']).sum()   # this results in a weird shape
```

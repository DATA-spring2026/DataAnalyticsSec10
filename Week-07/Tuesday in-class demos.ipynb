## Continuing with NumPy

### Briefly, review of hstack / vstack / dstack

```python
import numpy as np

a = np.array([0, 1, 2, 3, 11, 12])
b = np.array([4, 5, 6, 7, 13, 14])

print(np.hstack((a,b)))  # equivalent to concatenatating on the 0 axis
print(np.vstack((a,b)))  # equivalent to concatenating on the 1 axis
print(np.dstack((a,b)))  # equivalent to concatenating on the 2 axis
```

```python
print(np.concatenate((a,b)))

print(np.stack((a,b)))

print(np.stack((a,b), axis = 1))   # not equivalent to dstack, as dstack will return 3D array
```

```python
# so splitting arrays is breaking arrays apart

print(np.array_split(a, 2))
```

```python
print(np.array_split(a, 3))
```

```python
print(np.array_split(a, 4))  # allows for an uneven split
```

```python
print(np.hsplit(a, 2))  # hsplit works on the 0 axis, is not forgiving of uneven splits
```

```python
ab = np.stack((a,b))

print(ab)

print(np.vsplit(ab, 2))
```

```python
print(np.hsplit(ab, 2))
```

```python
# let's put this into context with some sales data

full_year = np.array([[100, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210, 220],
                      [200, 210, 190, 205, 220, 230, 240, 235, 225, 215, 205, 195],
                      [ 90,  95, 100, 110, 115, 120, 125, 130, 120, 115, 110, 105]])

# each array is a column
# the "rows" match on the vertical alignment
```

```python
quarters = np.array_split(full_year, 4, axis = 1)
print(quarters)

q1, q2, q3, q4 = quarters
print(q1)
```

```python
by_product = np.array_split(full_year, 3, axis = 0)

prod_a, prod_b, prod_c = by_product

print(prod_a)
```

### Using where() to apply a condition to an array

```python
print(np.where(full_year > 200))   # with only a condition, return an array of matches
```

```python
print(np.where(full_year > 200, 'high', full_year))  # lets you set a return and not-found value
```

```python
print(full_year[full_year > 200])  # filtering with bracket notation returns only matches
```

```python
print(full_year[full_year % 3 == 0])   # can put condition directly in brackets
```

```python
div_by_3 = (full_year % 3 == 0)

print(full_year[div_by_3])   # this is the pattern on p. 15 in the Student Guide
```

## Random in NumPy

```python
# when calling the random module in NumPy:
# np.random.[function name]
```

```python
some_ints = np.random.randint(100, size = (5))  # assumes a zero start == from 0 to 100
print(some_ints)
```

```python
more_ints = np.random.randint(50, 100, size = (5, 3))  # assumes a zero start == from 0 to 100
print(more_ints)
```

```python
print(np.random.rand(5, 3))
```

```python
# working with choice()

print(some_ints)  # starting values to choose from
print(np.random.choice(some_ints))
```

```python
sales_sample = np.random.choice(str_x, size = (4, 10))

print(sales_sample)
```

```python
# choice() can also be used to assign probability values

weighted_sales_sample = np.random.choice(str_x, size = (4, 10), 
                                         p = [.1, .2, .2, .2, .1, .05, .05, .05, .05])

print(weighted_sales_sample)
```

```python
# for starting values: 
# ['I-1456' 'InstaDerp' '110-XR' '#Zenith' '012' '014' '110' '145' '008']
# assigned p values:
#       .1          .2       .2        .2    .1   .05   .05   .05   .05
```

### Creating a distribution curve using normal() 

```python
x = np.random.normal(loc=5, scale=2, size=(30))

print(x)
```

## ufuncs in NumPy

```python
sale_price = np.array([2.99, 12.99, 3.50, 4.50, 17.99, 22.00])

discount_amount = np.array([0, 0, 3.50, .50, 3, 0])

print(sale_price + discount_amount)
```

```python
print(np.add(sale_price, discount_amount))
```

```python
print(np.multiply(sale_price, .8))

# print(np.multiply(sale_price, 1 - .2))  # another common format for calculating % discount
```

```python
print(np.round(np.multiply(sale_price, .8), 2))
```

```python
# round() applies "banker's rounding"
# rounding at 5 rounds towards the nearest even value

apple = np.array((.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5))

print(np.round(apple))

print(sum(np.round(apple)))


# numpy also has a function called around() ... which is just an alias for round()
```

```python
# compare to conventional round in Python

babble = [.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5]

print(round(sum(babble)))
```

## Using sum() / cumsum() / prod() / cumprod()

```python
# sum() adds up elements of an array

print(np.sum(apple))
```

```python
pear = np.random.rand(5, 3)
print(pear)
```

```python
print(np.sum(pear))
```

```python
print(np.sum(pear, axis=1))  
# since arrays can have multiple dimensions, you might want to sum on another axis
```

```python
print(np.cumsum(pear))
```

```python
print(np.cumsum(pear, axis=1))
```

```python
print(np.sum(pear, axis=1))
```

```python
# similarly, prod() gives all elements multiplied together
# and cumprod() will give a rolling product

print(np.prod(pear))
```

```python
print(np.prod(pear, axis=1))
```

```python
print(np.cumprod(pear))
```

```python
# diff() provides the difference in values (as in subtraction)
# between elements in an array

print(sale_price)

print(np.diff(sale_price))
```

```python
print(np.diff(sale_price, n=2))
```

```python
print(np.diff(sale_price, n=5))
```

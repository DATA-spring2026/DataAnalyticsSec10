### using the random library

```python
import random
```

```python
a = random.randint(1, 100)
a
```

```python
print(a)
```

```python
b = []

for i in range(10):
    b.append(random.randint(1, 100))

print(b)
```

```python
random.choice(b)
```

```python
random.choices(b)
```

```python
# .choices() using optional parameters

random.choices(b, weights=[10,2,1,2,1,2,1,2,6,7], k=12)
```

```python
# .choices() using optional parameters
# parameter k can specify more than the number of values in the original target collection

random.choices(b, weights=[10,2,1,2,1,2,1,2,6,7], k=12)
```

```python
print(b)
```

```python
random.shuffle(b)
```

```python
print(b)
```

```python
# no weighting component to .sample()
# parameter k limited to maximum of the number of values in the original target collection

random.sample(b, 5)
```


 ## Custom Functions

 ### working with def keyword

```python
# very simple example of a function:

def pizza():
    return "Yay pizza!"
```

```python
pizza()
```

```python
def pizza_v_burritos(food_name):
    if food_name == 'pizza':
        msg = 'Yay pizza!'
    elif food_name == 'burritos':
        msg = 'Yay burritos!'
    else:
        msg = 'boooooooooo'
    return msg
```

```python
pizza_v_burritos('burritos')
```

```python
pizza_v_burritos('pizza')
```

```python
pizza_v_burritos('liver and onions')
```

```python
# pizza_v_burritos('pizza', 'burritos')
# will result in error -- function only accepts one argument
```

```python
# example of a function with two parameters

def place_order(item, qty):
    return f'You ordered {qty} of item {item}'
```

```python
place_order('pizza', 2)
```

```python
place_order('hockey puck', 67)
```

```python
place_order(input('What would you like to order (pizza or burrito)? '), input('How many? '))
```

## using keyword arguments

```python
# def place_order(item, qty):
#    return f'You ordered {qty} of item {item}'
```

```python
place_order(qty="7", item="burritos")
```

```python
def tax_calc(rate, amount, exempt='N'):
    if exempt != 'N':
        tax_amt = 0
    else:
        tax_amt = rate * amount
    return tax_amt
```

```python
tax_calc(.08, 122.45)
```

```python
tax_calc(.075, 58, 'exempt')
```

## working with an arbitrary number of arguments

```python
def greeting_kids(*args):
    return args
```

```python
greet = greeting_kids('Henry','Alice')

greet
```

```python
def send_greetings(whomst):
    for i in whomst:
        print(f'Hi, {i}!')
```

```python
send_greetings(greet)
```

```python
send_greetings(['Bob', 'Nancy', 'Charlemagne'])
```

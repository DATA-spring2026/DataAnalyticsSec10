## More function examples

```python
# .find() searches left to right

order_status = "order #P080624 is ready for pickup"
print(order_status.find("is ready"))
print(order_status.find("p"))
```

```python
order_status = "order #P080624 is ready for pickup"
print(order_status.find("orange"))   # value not found gives -1
```

```python
muppet = 'miss piggy'
print(muppet.find('y'))
```

```python
# rfind reads right to left

order_status = "order #P080624 is ready for pickup"
print(order_status.rfind("is ready"))
print(order_status.rfind("p"))
```

```python
order_status.find("P0")
```

```python
# tangent about using print() versus just calling a variable by name in Notebook

bank_balance = 100

print(bank_balance)
print(order_status)
```

```python
order_status[order_status.find("P0")]
# gets the same result as order_status[7]
```

```python
order_status[order_status.find("P0"):order_status.find("P0")+7]
# gets the same result as a range of order_status[7:7+7]
```

```python
order_status[7:7+7]
```

```python
tues_orders = ["order #P080624 is ready for pickup", "order in progress #P080626", "customer picked up order P080623", "P080620 was cancelled"]

orders = []

for i in tues_orders:
    if "cancelled" not in i:
        orders.append(i[i.find("P0"):i.find("P0")+7])

print(orders)
```

## String formatting

```python
# example from the book

txt1 = "The ribeye is {price:.2f} dollars, yikes"
print(txt1.format(price = 49))
```

```python
# expanded example

final_receipt = "Total plus tax is ${total:.2f}"

food_bill = {"entree1":23, "entree2":18, "drink1":4.00, "drink2":7.00}

subtotal = sum(food_bill.values())

tax = subtotal * .08

print(final_receipt.format(total = subtotal + tax))
```
